$
ci(xy) = - \int_x^{\infty} \frac{cos(zy)}{z}dz
$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

# 定义公式
def integrand(z, y):
    return -np.cos(z * y) / z

def ci(x, y):
    result, _ = quad(integrand, x, np.inf, args=(y,))
    return result

# 创建一个网格
x = np.linspace(0.1, 5, 50)
y = np.linspace(0.1, 5, 50)
X, Y = np.meshgrid(x, y)
Z_real = np.zeros_like(X)
Z_imag = np.zeros_like(X)

# 计算ci(xy)的实部和虚部
for i in range(len(x)):
    for j in range(len(y)):
        Z = ci(X[i, j], Y[i, j])
        Z_real[i, j] = np.real(Z)
        Z_imag[i, j] = np.imag(Z)

# 创建3D图形
fig = plt.figure(figsize=(12, 6))

# 子图1 - 实部
ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z_real, cmap='viridis')
ax1.set_title('ciReal Part')

# 子图2 - 虚部
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z_imag, cmap='viridis')
ax2.set_title('Imaginary Part')

# 添加颜色条
fig.colorbar(surf1, ax=ax1)
fig.colorbar(surf2, ax=ax2)

# 设置轴标签
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Real(ci(xy))')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Imag(ci(xy))')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
