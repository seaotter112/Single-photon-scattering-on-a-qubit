$ 
R(\omega_s,x,t)=R(\omega_s)e^{-i\frac{\omega_s}{v_g}(x+v_gt)}$   
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;
$+\frac{R(\omega_s)}{2\pi}$ $e^{-i\frac {\omega_s}{v_g}(x+v_gt)}[ici(\omega_{s}\frac{|x|}{v_g})+si(\omega_s\frac{|x|}{v_g})]$  
$Eq(44)$  
&emsp; &emsp; &emsp; &emsp; $x<0$ &emsp; $x+v_gt>0$  

$
ci(xy) = - \int_x^{\infty} \frac{cos(zy)}{z}dz
$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

# 定义公式
def integrand(z, y):
    return -np.cos(z * y) / z

def ci(x, y):
    result, _ = quad(integrand, x, np.inf, args=(y,))
    return result

# 创建一个网格
#x = np.linspace(0.995*5*2*math.pi, 1.005*5*2*math.pi, 50) # from Eq(44) FIG.6 設定參數 omega_s
#y = np.linspace(0.001/(3*10**8), 0.001/(3*10**8), 50) # from Eq(44) FIG.6 設定參數
x = np.linspace(0.1, 5, 50)
y = np.linspace(0.1, 5, 50)
X, Y = np.meshgrid(x, y)
Z_real = np.zeros_like(X)
Z_imag = np.zeros_like(X)

# 计算ci(xy)的实部和虚部
for i in range(len(x)):
    for j in range(len(y)):
        Z = ci(X[i, j], Y[i, j])
        Z_real[i, j] = np.real(Z)
        Z_imag[i, j] = np.imag(Z)

# 创建3D图形
fig = plt.figure(figsize=(12, 6))

# 子图1 - 实部
ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z_real, cmap='viridis')
ax1.set_title('ci Real Part')

# 子图2 - 虚部
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z_imag, cmap='viridis')
ax2.set_title('ci Imaginary Part')

# 添加颜色条
fig.colorbar(surf1, ax=ax1)
fig.colorbar(surf2, ax=ax2)

# 设置轴标签
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Real(ci(xy))')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Imag(ci(xy))')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

$
si(xy) = - \int_x^{\infty} \frac{sin(zy)}{z}dz
$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

# 定义新的公式
def integrand(z, x, y):
    return np.sin(z * y) / z

def si(x, y):
    result, _ = quad(integrand, x, np.inf, args=(x, y))
    return -result

# 创建一个网格
# x = np.linspace(0.995*5*2*math.pi, 1.005*5*2*math.pi, 50) # from Eq(44) FIG.6 設定參數 omega_s
# y = np.linspace(0.001/(3*10**8), 0.001/(3*10**8), 50) # from Eq(44) FIG.6 設定參數
x = np.linspace(0.1, 5, 50)
y = np.linspace(0.1, 5, 50)
X, Y = np.meshgrid(x, y)
Z_real = np.zeros_like(X)
Z_imag = np.zeros_like(X)

# 计算si(xy)的实部和虚部
for i in range(len(x)):
    for j in range(len(y)):
        Z = si(X[i, j], Y[i, j])
        Z_real[i, j] = np.real(Z)
        Z_imag[i, j] = np.imag(Z)

# 创建3D图形
fig = plt.figure(figsize=(12, 6))

# 子图1 - 实部
ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z_real, cmap='viridis')
ax1.set_title('si Real Part')

# 子图2 - 虚部
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z_imag, cmap='viridis')
ax2.set_title('si Imaginary Part')

# 添加颜色条
fig.colorbar(surf1, ax=ax1)
fig.colorbar(surf2, ax=ax2)

# 设置轴标签
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Real(si(xy))')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Imag(si(xy))')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
